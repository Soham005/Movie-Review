```python
import pandas as pd
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
```

```python
data = pd.read_csv('IMDB Dataset-2.csv')
```

```python
data
```

```python
data['sentiment'].value_counts()
```

```python
data.dropna(inplace=True)
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def clean_review(review):
    return ' '.join(word for word in review.split() if word.lower() not in stop_words)

data['text'] = data['text'].apply(clean_review)
```

```python
#wordcloud 

from wordcloud import WordCloud
```

```python
#draw WordCloud
reviews = ' '.join(word for word in data['text'][data['sentiment'] == 'neg'].astype(str))
wordcloud = WordCloud(height = 600, width = 1000, max_font_size = 100)

plt.figure(figsize = (15,12))
plt.imshow(wordcloud.generate(reviews), interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

```python
reviews = ' '.join(word for word in data['text'][data['sentiment'] == 'pos'].astype(str))
wordcloud = WordCloud(height = 600, width = 1000, max_font_size = 100)

plt.figure(figsize = (15,12))
plt.imshow(wordcloud.generate(reviews), interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(max_features=2500)
```

```python
reviews = cv.fit_transform(data['text']).toarray()
```

```python
data['sentiment'] = data['sentiment'].replace(['pos','neg'],[1,0])
```

```python
data['sentiment'].value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
reviews_train, reviews_test, sent_train, sent_test = train_test_split(reviews, data['sentiment'], test_size = 0.2)
```

```python
model.fit(reviews_train, sent_train)
```

```python
predict = model.predict(reviews_test)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm = confusion_matrix(sent_test, predict, labels=model.classes_)
display = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels=model.classes_)
display.plot()
plt.show()
```

```python
import pickle as pk
```

```python
pk.dump(model,open('model.pkl','wb'))
```

```python
pk.dump(cv,open('scaler.pkl','wb'))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(reviews_train, sent_train)
sent_pred_lr = log_reg.predict(reviews_test)
```

```python
print("Logistic Regression Results:")
print(classification_report(sent_test, sent_pred_lr))
print("Accuracy:", accuracy_score(sent_test, sent_pred_lr))
print("-"*60)
```

```python
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)
rf_clf.fit(reviews_train, sent_train)
sent_pred_rf = rf_clf.predict(reviews_test)
```

```python
print("Random Forest Results:")
print(classification_report(sent_test, sent_pred_rf))
print("Accuracy:", accuracy_score(sent_test, sent_pred_rf))
```
